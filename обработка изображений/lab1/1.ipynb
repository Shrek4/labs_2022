```python
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
import numpy as np
import math as m
import cv2
import matplotlib.pylab as pl
from skimage.exposure import histogram
```

```python
fig = plt.figure()

fig, (ax1, ax2 ) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 10)
)

ax1.set_title('01')
ax2.set_title('')


image = cv2.imread("test.jpg") # . 


ax1.imshow(image)
ax2.hist(image.ravel(),bins = 256, histtype = 'step' )

plt.show()

print("  :\t", np.max(image))
print("  :\t", np.min(image))
print(":\t", np.max(image) - np.min(image))
```

```python
# 
fig = plt.figure()

fig, [ax1, ax2 ] = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 10)
)

ax1.set_title('01_log')
ax2.set_title('')

def log_p(img):
    c=int(255/(np.log1p(1+np.max(img))))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for p in range(len(img[i][j])):
                img[i][j][p] = c * np.log1p(1+img[i][j][p])

log_p(image)
ax1.imshow(image)
ax2.hist(image.ravel(),bins = 256, histtype = 'step' )
plt.show()
print("  :\t", np.max(image))
print("  :\t", np.min(image))
print(":\t", np.max(image) - np.min(image))
```

```python
#
fig = plt.figure()

fig, [ax1, ax2 ] = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 10)
)

ax1.set_title('01_log')
ax2.set_title('')

def deg(img, pow, c = 1):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for p in range(len(img[i][j])):
                img[i][j][p] = 255 * (img[i][j][p]/255) **pow

deg(image, 1.5)
ax1.imshow(image)
ax2.hist(image.ravel(),bins = 256, histtype = 'step' )
plt.show()
print("  :\t", np.max(image))
print("  :\t", np.min(image))
print(":\t", np.max(image) - np.min(image))
```

```python
#
fig = plt.figure()

fig, (ax1, ax2 ) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 10)
)

ax1.set_title('')
ax2.set_title('')

image1N = image.copy()

image1N = cv2.normalize(image1N, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

ax1.imshow(image1N)
ax2.hist(image1N.ravel(), bins = 256, histtype = 'step')

plt.show()

print("  :\t", np.max(image1N))
print("  :\t", np.min(image1N))
print(":\t", np.max(image1N) - np.min(image1N))
```

```python
#
fig = plt.figure()

fig, (ax1, ax2 ) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(10, 10)
)

ax1.set_title('')
ax2.set_title('')

imageQ = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imageQ = cv2.equalizeHist(imageQ)

ax1.imshow(imageQ, cmap='gray')
ax2.hist(imageQ.ravel(), bins = 256, histtype = 'step')

plt.show()

print("  :\t", np.max(imageQ))
print("  :\t", np.min(imageQ))
print(":\t", np.max(imageQ) - np.min(imageQ))
```
